```python
ensembel = pd.read_csv(f"{DATA_PATH}ensemble.csv")
sj_1 = pd.read_csv(f"{DATA_PATH}LGBM_submission_1.csv")
sj_2 = pd.read_csv(f"{DATA_PATH}LGBM_submission_2.csv")
sw = pd.read_csv(f"{DATA_PATH}submission_1_100_nsw_ver4.csv")
ys = pd.read_csv(f"{DATA_PATH}submission_feature5 (3).csv")
sj_3 = pd.read_csv(f"{DATA_PATH}LGBM_submission_csj.csv")
sw_2 = pd.read_csv(f"{DATA_PATH}submission_1_100_nsw_ver9.csv")
ys_2 = pd.read_csv(f"{DATA_PATH}submission_feature7 (2).csv")
```

```python
ensembel_2 = pd.concat([ensembel, sj_1['answer'], sj_2['answer'], sw['answer'], ys['answer'], sw_2['answer'], ys_2['answer']], axis=1)
ensembel_2
```

```python
cols = ['num_date_time', 'answer', 'sub_1', 'sub_2', 'sub_3', 'sub_4', 'sub_5', 'sub_6']
ensembel_2.columns = cols
ensembel_2
```

```python
ensembel_2["근사치"] = ensembel_2.apply(lambda row: min(row["sub_1"], row["sub_2"], row["sub_3"], row["sub_4"], row["sub_5"], row["sub_6"], key=lambda x: abs(x - row['answer'])), axis=1)
```

```python
ensembel_2
```

```python
submission = pd.read_csv(f"{DATA_PATH}sample_submission.csv")
```

```python
submission['answer'] = (ensembel_2['answer'] + ensembel_2['근사치']) / 2
submission
```

```python
submission.to_csv(f"{DATA_PATH}ensemble_2.csv", index=False)
```
