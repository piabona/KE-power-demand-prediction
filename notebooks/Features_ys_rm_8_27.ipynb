```python
train_RM = train_df[['date_time', 'building_num', 'temp', 'wind', 'hum']] # 'pred'는 실제값 사용
test_RM = test_df[['date_time', 'building_num', 'temp', 'wind', 'hum']] # 'pred'는 실제값 사용
```

```python
rolling_mean_df = pd.concat([train_RM,test_RM], axis=0).reset_index(drop=True)
rolling_mean_df.shape , rolling_mean_df.isnull().sum().sum()
```

```python
# 이동평균 12시간
window_size = 12
rolling_mean_df['temp_RM'] = rolling_mean_df.groupby('building_num')['temp'].rolling(window=window_size).mean().reset_index(level=0, drop=True)
# rolling_mean_df['pred_RM'] = rolling_mean_df.groupby('building_num')['pred'].rolling(window=window_size).mean().reset_index(level=0, drop=True)
rolling_mean_df['wind_RM'] = rolling_mean_df.groupby('building_num')['wind'].rolling(window=window_size).mean().reset_index(level=0, drop=True)
rolling_mean_df['hum_RM'] = rolling_mean_df.groupby('building_num')['hum'].rolling(window=window_size).mean().reset_index(level=0, drop=True)
rolling_mean_df
```

```python
# 지수이동평균 12시간
window_size = 12
rolling_mean_df['temp_EMA'] = rolling_mean_df.groupby('building_num')['temp'].transform(lambda x: x.ewm(span=window_size, adjust=False).mean())
# rolling_mean_df['pred_EMA'] = rolling_mean_df.groupby('building_num')['pred'].transform(lambda x: x.ewm(span=window_size, adjust=False).mean())
rolling_mean_df['wind_EMA'] = rolling_mean_df.groupby('building_num')['wind'].transform(lambda x: x.ewm(span=window_size, adjust=False).mean())
rolling_mean_df['hum_EMA'] = rolling_mean_df.groupby('building_num')['hum'].transform(lambda x: x.ewm(span=window_size, adjust=False).mean())
rolling_mean_df
```

```python
train_df['THI'] = 1.8*train_df['temp_RM'] - 0.55*(1-train_df['hum_RM']/100)*(1.8*train_df['temp_RM']-26)+32
test_df['THI'] = 1.8*test_df['temp_RM'] - 0.55*(1 - test_df['hum_RM']/100)*(1.8*test_df['temp_RM'] - 26)+32
```

```python
# 체감온도 변수 추가
def wet_temp(temp, hum):
    tw = temp*np.arctan(0.151977*(hum + 8.313659)*0.5) + np.arctan(temp + hum) - np.arctan(hum - 1.67633) + 0.00391838*1.5*np.arctan(0.023101*hum) - 4.686035
    return tw

def feels_temp(temp, hum):
    tw = wet_temp(temp, hum)
    ft = -0.2442 + 0.55399*tw + 0.45535*temp - 0.0022*(tw**2) + 0.00278*tw*temp + 3.0
    return ft
```

```python
train_df['feels_temp'] = feels_temp(train_df['temp_RM'], train_df['hum_RM'])
test_df['feels_temp'] = feels_temp(test_df['temp_RM'], test_df['hum_RM'])
```

```python
# cdh(cooling degree hour) 변수 추가
def CDH(xs):
    ys = []
    for i in range(len(xs)):
        if i < 11:
            ys.append(np.sum(xs[:(i+1)]-26))
        else:
            ys.append(np.sum(xs[(i-11):(i+1)]-26))
    return np.array(ys)
```

```python
cdhs = np.array([])
for num in range(1,101,1):
    temp = train_df[train_df['building_num'] == num]
    cdh = CDH(temp['temp_RM'].values)
    cdhs = np.concatenate([cdhs, cdh])
train_df['CDH'] = cdhs

cdhs = np.array([])
for num in range(1,101,1):
    temp = test_df[test_df['building_num'] == num]
    cdh = CDH(temp['temp_RM'].values)
    cdhs = np.concatenate([cdhs, cdh])
test_df['CDH'] = cdhs
```

```python
# 시간별 전력량 평균 추가
power_hour_mean = train_df.groupby(['building_num', 'time'])['power'].mean().reset_index()
train_df['hour_mean'] = train_df.apply(lambda x :power_hour_mean.loc[(power_hour_mean.building_num == x['building_num']) & (power_hour_mean.time == x['time']), 'power'].values[0], axis=1)
test_df['hour_mean'] = test_df.apply(lambda x :power_hour_mean.loc[(power_hour_mean.building_num == x['building_num']) & (power_hour_mean.time == x['time']), 'power'].values[0], axis=1)
```

```python
# 요일별 시간별 전력량 평균 추가
power_mean = pd.pivot_table(train_df, values = 'power', index = ['building_num', 'time', 'day'], aggfunc = np.mean).reset_index()

train_df['day_hour_mean'] = train_df.apply(lambda x : power_mean.loc[(power_mean.building_num == x['building_num']) & (power_mean.time == x['time']) & (power_mean.day == x['day']) ,'power'].values[0], axis = 1)
test_df['day_hour_mean'] = test_df.apply(lambda x : power_mean.loc[(power_mean.building_num == x['building_num']) & (power_mean.time == x['time']) & (power_mean.day == x['day']) ,'power'].values[0], axis = 1)
```

```python
# 시간별 전력량 표준편차 추가
power_hour_std = pd.pivot_table(train_df, values = 'power', index = ['building_num', 'time'], aggfunc = np.std).reset_index()

train_df['hour_std'] = train_df.apply(lambda x : power_hour_std.loc[(power_hour_std.building_num == x['building_num']) & (power_hour_std.time == x['time']) ,'power'].values[0], axis = 1)
test_df['hour_std'] = test_df.apply(lambda x : power_hour_std.loc[(power_hour_std.building_num == x['building_num']) & (power_hour_std.time == x['time']) ,'power'].values[0], axis = 1)
```

```python
# 열지수 단계별로 구분
T = train_df['temp_RM']*9/5 + 32 # 화씨로 전환
RH = train_df['hum_RM']
train_df['heat_index'] = -42.379 + 2.04901523*T + 10.14333127*RH - .22475541*T*RH - .00683783*T*T - .05481717*RH*RH + .00122874*T*T*RH + .00085282*T*RH*RH - .00000199*T*T*RH*RH
train_df['heat_index'] = (train_df['heat_index']-32) * 5/9 # 섭씨로 전환
train_df.loc[train_df['heat_index']<32, 'heat_index'] = 0
train_df.loc[(train_df['heat_index']>=32) & (train_df['heat_index']<41), 'heat_index'] = 1
train_df.loc[(train_df['heat_index']>=41) & (train_df['heat_index']<54), 'heat_index'] = 2
train_df.loc[(train_df['heat_index']>=54) & (train_df['heat_index']<66), 'heat_index'] = 3
train_df.loc[train_df['heat_index']>=66, 'heat_index'] = 4

T = test_df['temp_RM']*9/5 + 32 # 화씨로 전환
RH = test_df['hum_RM']
test_df['heat_index'] = -42.379 + 2.04901523*T + 10.14333127*RH - .22475541*T*RH - .00683783*T*T - .05481717*RH*RH + .00122874*T*T*RH + .00085282*T*RH*RH - .00000199*T*T*RH*RH
test_df['heat_index'] = (test_df['heat_index']-32) * 5/9 # 섭씨로 전환
test_df.loc[test_df['heat_index']<32, 'heat_index'] = 0
test_df.loc[(test_df['heat_index']>=32) & (test_df['heat_index']<41), 'heat_index'] = 1
test_df.loc[(test_df['heat_index']>=41) & (test_df['heat_index']<54), 'heat_index'] = 2
test_df.loc[(test_df['heat_index']>=54) & (test_df['heat_index']<66), 'heat_index'] = 3
test_df.loc[test_df['heat_index']>=66, 'heat_index'] = 4
```

- 7일전 대비 전력 변화량의 평균 (168일)

```python
train_df['power_change_rate_7d'] = (train_df['power'] - train_df.groupby('building_num')['power'].shift(7*24)) / train_df.groupby('building_num')['power'].shift(7*24)

train_df['day_within_cycle'] = train_df.index % (7*24)

avg_power_change_rate_7d = train_df.groupby(['building_num','day_within_cycle' ])['power_change_rate_7d'].mean().reset_index()
avg_power_change_rate_7d.rename(columns={'power_change_rate_7d': 'avg_power_change_rate_7d'}, inplace=True)

train_df = train_df.merge(avg_power_change_rate_7d, on=['building_num','day_within_cycle'], how='left')
cols = ['day_within_cycle','power_change_rate_7d']
train_df = train_df.drop(columns = cols)

train_df['avg_power_change_rate_7d'].nunique()

last_rows = train_df.groupby('building_num').tail(168)['avg_power_change_rate_7d'].reset_index().drop('index',axis=1)
test_df = pd.concat([test_df, last_rows], axis=1)
```

- 1일전 대비 전력 변화량의 평균 (24시간)

```python
train_df['power_change_rate_24h'] = (train_df['power'] - train_df.groupby('building_num')['power'].shift(24)) / train_df.groupby('building_num')['power'].shift(24)

train_df['day_within_cycle'] = train_df.index % (24)

avg_power_change_rate_24h = train_df.groupby(['building_num','day_within_cycle' ])['power_change_rate_24h'].mean().reset_index()
avg_power_change_rate_24h.rename(columns={'power_change_rate_24h': 'avg_power_change_rate_24h'}, inplace=True)

train_df = train_df.merge(avg_power_change_rate_24h, on=['building_num','day_within_cycle'], how='left')
cols = ['day_within_cycle','power_change_rate_24h']
train_df = train_df.drop(columns = cols)

last_rows = train_df.groupby('building_num').tail(168)['avg_power_change_rate_24h'].reset_index().drop('index',axis=1)
test_df = pd.concat([test_df, last_rows], axis=1)
```
